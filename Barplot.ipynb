```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv(r"C:\Users\sadia\OneDrive\Desktop\ML\iris.csv")
```

```python
iris_means = iris.groupby('species').mean()
iris_means = iris_means.reset_index()
iris_melted = iris_means.melt(id_vars="species", var_name="feature", value_name="mean_value")
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x="feature", y="mean_value", hue="species", data=iris_melted)
plt.title("Average Feature Value by Species")
plt.show()
```
